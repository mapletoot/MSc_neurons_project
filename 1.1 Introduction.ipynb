# Libraries

```python
# Import packages for data manipulation and visualization
import pandas as pd
import anndata as ad
import numpy as np

# Import packages for neuron visualizations

from brainrender import Scene
from brainrender.actors import Points
import vedo
vedo.settings.default_backend= 'vtk'
```

```python
# load data
ec_snr_data = pd.read_csv("Data/master_detailed_comment.csv")

# load EC data (imputed MERFISH) object
source = "Data/ec_obj_imputed_log2.h5ad"
ec_data = ad.read_h5ad(source)
cell_metadata = ec_data.obs # Extract metadata that contains layer type
```

```python

aec_cells = cell_metadata[cell_metadata["substructure"].isin(["ENTm1", "ENTl1"])]
aec_coords = np.array(aec_cells[["x_ccf", "y_ccf", "z_ccf"]])
aec_points = Points(aec_coords, radius=20, colors="red", alpha=0.05)

bec_cells = cell_metadata[cell_metadata["substructure"].isin(["ENTm2", "ENTl2"])]
bec_coords = np.array(bec_cells[["x_ccf", "y_ccf", "z_ccf"]])
bec_points = Points(bec_coords, radius=20, colors="orange", alpha=0.05)

cec_cells = cell_metadata[cell_metadata["substructure"].isin(["ENTm3", "ENTl3"])]
cec_coords = np.array(cec_cells[["x_ccf", "y_ccf", "z_ccf"]])
cec_points = Points(cec_coords, radius=20, colors="yellow", alpha=0.05)

dec_cells = cell_metadata[cell_metadata['substructure'].isin(["ENTm5", "ENTl5"])]
dec_coords = np.array(dec_cells[["x_ccf", "y_ccf", "z_ccf"]])
dec_points = Points(dec_coords, radius=20, colors="green", alpha=0.05)

eec_cells = cell_metadata[cell_metadata["substructure"].isin(["ENTm6", "ENTl6a"])]
eec_coords = np.array(eec_cells[["x_ccf", "y_ccf", "z_ccf"]])
eec_points = Points(eec_coords, radius=20, colors="blue", alpha=0.05)


xec_cells = ec_snr_data
xec_coords = np.array(xec_cells[["x", "y", "z"]])
xec_points = Points(xec_coords, radius=30, colors="black")


scene = Scene(
    atlas_name="allen_mouse_25um",
    title="EC Neurons (Coloured by Layer)"
)

scene.add(aec_points)
scene.add(bec_points)
scene.add(cec_points)
scene.add(dec_points)
scene.add(eec_points)
scene.add(xec_points)

scene.add_brain_region("ENTl", alpha=0.1, color="blue")
scene.add_brain_region("ENTm", alpha=0.1, color="blue")

# cam = scene.plotter.camera

# cam.SetPosition(6610.21, -31614.6, -6569.89)
# cam.SetFocalPoint(7829.74, 4296.03, -5694.5)
# cam.SetViewUp(-0.999424, 0.0339266, 0.000575096)
# cam.SetClippingRange(25503.4, 46935.6)
# cam.SetViewAngle(18.75)

scene.render(camera='top')
scene.plotter.screenshot("Figures/fig_1_top_view.png")
```
